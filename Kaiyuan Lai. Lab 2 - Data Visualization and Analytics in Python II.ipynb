```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('Superstore.csv')

df
```

```python
plt.figure(figsize=(8,4))

ax = sns.barplot(x='Category', y='Sales', data=df, estimator=np.mean)

ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
```

```python
plt.figure(figsize=(8,4))

sns.countplot(data=df, x="State", order=df["State"].value_counts().index)
plt.xticks(rotation=90)

plt.title("Transaction Count by State")
plt.xlabel("State")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
df["Year"] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y').dt.year

plt.figure(figsize=(8,4))

ax = sns.barplot(x='Year', y='Sales', data=df, estimator=np.mean)

ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)

plt.title("average sales by year")
plt.xlabel("year")
plt.ylabel("Number of sales")
plt.show()
```

```python
plt.figure(figsize=(8,4))

# create boxplot in seaborn with a color palette
sns.boxplot(x="Region", y="Shipping Cost", data=df, palette='rainbow')
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x="Region", y="Shipping Cost", hue="Ship Mode", data=df, palette="rainbow")
```

```python
g = sns.jointplot(
    data=df,
    x="Shipping Cost",
    y="Unit Price",
    hue="Container",
    kind="scatter",   # scatterplot in the middle
    height=8
)

# Set axis limits
g.ax_joint.set_xlim(0, 40)
g.ax_joint.set_ylim(0, 100)

# Add title
plt.suptitle("Shipping Cost vs. Unit Price by Container", y=1.02)

mask = (df["Shipping Cost"] >= 18) & (df["Shipping Cost"] <= 22)
container_counts = df.loc[mask, "Container"].value_counts()
most_common_container = container_counts.idxmax()
print(container_counts)
print(f"\nThe container used most often for shipping costs around $20 is: {most_common_container}")
```

```python
cols = ["Unit Price", "Sales", "Profit", "Shipping Cost"]

# Compute correlation matrix
corr = df[cols].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, fmt=".2f")
```

```python
 df.groupby(["Category", "Region"])["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False)
```

```python
df_postcode = df.groupby(["Postal Code"])["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False)

df_postcode[df_postcode['Sales'] > 100000].count()
```

```python
df_trasactionvalue = df.groupby('Category')['Sales'].max().sort_values(ascending = False)

df_trasactionvalue.loc['Envelopes']
```

```python
df_central = df[df['Region'].str.contains('Central')]

df_central.groupby('Item ID')['Order Quantity'].sum().sort_values(ascending = False)
```

```python
df['Profit_Margin'] = df['Profit'] / df['Sales']

df.groupby('Item ID')['Profit_Margin'].mean().sort_values(ascending = False)
```

```python
customer_sales = df.groupby("Customer ID")["Sales"].sum().reset_index()

# Step 2: Sort by total sales descending
customer_sales_sorted = customer_sales.sort_values(by="Sales", ascending=False).reset_index(drop=True)

# --- Method 1: Using iloc ---
customer_id_iloc = customer_sales_sorted.iloc[99]["Customer ID"]   # 100th row (0-based index)

# --- Method 2: Using rank ---
customer_sales["Rank"] = customer_sales["Sales"].rank(method="min", ascending=False)
customer_id_rank = customer_sales.loc[customer_sales["Rank"] == 100, "Customer ID"].values[0]

print("Method 1 (iloc):", customer_id_iloc)
print("Method 2 (rank):", customer_id_rank)
```

```python
df["Month"] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y').dt.month

df.groupby(["Year", "Month"])["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False).head(1)
```
