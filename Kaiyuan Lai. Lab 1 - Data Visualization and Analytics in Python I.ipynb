```python
df4 = df.value_counts('Region').to_frame().reset_index()
df4.columns = ['Region', 'number']

fig, ax = plt.subplots(figsize=(8,6))

ax.pie(df4.number, labels = df4.Region, autopct ='% .2f%%', startangle = 90)

ax.set_title('percentage of region')

plt.show()
```
